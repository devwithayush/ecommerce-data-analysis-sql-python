## Missing Values Check

```python
def check_missing_values(df, name):
    print(f"\n{name}")
    print("-" * 60)
    
    missing = df.isnull().sum()
    missing_percent = (df.isnull().sum() / len(df)) * 100
    
    missing_df = pd.DataFrame({
        'Column': df.columns,
        'Missing_Count': missing.values,
        'Missing_Percent': missing_percent.values
    })
    
    missing_with_nulls = missing_df[missing_df['Missing_Count'] > 0]
    
    if len(missing_with_nulls) == 0:
        print("--> No missing values")
    else:
        print(missing_with_nulls.to_string(index=False))

check_missing_values(customers, "CUSTOMERS")
check_missing_values(orders, "ORDERS")
check_missing_values(order_items, "ORDER ITEMS")
check_missing_values(payments, "PAYMENTS")
check_missing_values(reviews, "REVIEWS")
check_missing_values(products, "PRODUCTS")
check_missing_values(sellers, "SELLERS")
check_missing_values(geolocation, "GEOLOCATION")
check_missing_values(category_translation, "CATEGORY TRANSLATION")
```

<hr>

## Duplicates Check

```python
def check_duplicates(df, name, key_column):
    print(f"\n{name}")
    print("-" * 60)
    
    total_duplicates = df.duplicated().sum()
    print(f"Total duplicate rows (entire row): {total_duplicates}")
    
    if key_column:
        key_duplicates = df[key_column].duplicated().sum()
        print(f"Duplicate {key_column}: {key_duplicates}")

check_duplicates(customers, "CUSTOMERS", 'customer_id')
check_duplicates(orders, "ORDERS", 'order_id')
check_duplicates(order_items, "ORDER ITEMS", 'order_id') 
check_duplicates(payments, "PAYMENTS", 'order_id')  
check_duplicates(reviews, "REVIEWS", 'review_id')
check_duplicates(products, "PRODUCTS", 'product_id')
check_duplicates(sellers, "SELLERS", 'seller_id')
check_duplicates(geolocation, "GEOLOCATION", 'geolocation_zip_code_prefix')
check_duplicates(category_translation, "CATEGORY TRANSLATION", 'product_category_name')
```

<br>
Duplicate rows were checked across all datasets.

- Customers, orders, products, sellers, and category translation datasets do not have duplicate IDs, which is expected for primary keys.

- order_items and payments contain duplicate order_id values. This is normal because:
  - One order can have multiple items
  - One order can have multiple payment records

- Geolocation dataset contains a large number of duplicate rows and repeated zip code prefixes. This is expected since multiple entries can exist for the same location.

- Reviews dataset has duplicate review_id values. This is not expected and may indicate data inconsistency or duplicate entries or Maybe it's allow in this dataset that one person can give multiple review . This will be investigated further during the cleaning phase.

Overall, most duplicates are expected based on dataset structure, except for the review_id duplicates which need further checking.
